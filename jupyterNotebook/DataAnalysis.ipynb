# Data Cleaning



Cleaning Data is essential for analysis

### Converting Numeric Features to numeric datatype

#### Cleaning column Ex-Showroom_Price

```python
cars["Ex-Showroom_Price"]=cars["Ex-Showroom_Price"].str.replace('R','').str.replace('s','')
```

```python
cars["Ex-Showroom_Price"]=cars["Ex-Showroom_Price"].str.replace('[\,\,\.]','').astype(int)
```

```python
cars
```

#### Cleaning column Displacement

```python
cars["Displacement"]=cars["Displacement"].str.replace('c','').astype(float)
```

#### Cleaning columns Height, Width, Length, Wheelbase

```python
cleanedColumns=["Height","Width","Length","Wheelbase"]
for col in cleanedColumns:
    cars[col]=cars[col].str.replace('m','').astype(float)
    
```

#### Cleaning columns City_Mileage, ARAI_Certified_Mileage

```python
cleanColumns=["City_Mileage","ARAI_Certified_Mileage"]
for col in cleanColumns:
    print(cars[col].dtype)
    cars[col]=cars[col].str.replace('k','').str.replace('m','').str.replace('/','').str.replace('l','').str.replace('i','').str.replace('t','').str.replace('s','').str.replace('e','').str.replace('r','').str.replace('?','').str.replace(',','').str.replace('p','').str.replace(' ','')
    cars[col]=cars[col].str.partition('-')[0]
    cars[col]=cars[col].astype(float)
                                                                             
```

#### Cleaning column Kerb_Weight

```python
cars["Kerb_Weight"]=cars["Kerb_Weight"].str.replace('k','').str.replace('g','').str.partition('-')[0].astype(float)
```

#### Cleaning column Fuel_Tank_Capacity

```python
cars["Fuel_Tank_Capacity"]=cars["Fuel_Tank_Capacity"].str.replace('l','').str.replace('i','').str.replace('t','').str.replace('s','').str.replace('e','').str.replace('r','')
cars["Fuel_Tank_Capacity"]=cars["Fuel_Tank_Capacity"].astype(float)
```

### Handling NAN Values

#### Checking if there are null values in dataframe

```python
cars.isnull().sum()
```

```python
columns=cars.columns
columns
```

```python
cars["Ex-Showroom_Price"].dtype
```

#### Filling NAN of numeric features with there mean 

```python
for column in columns:
    if(cars[column].isnull().sum()!=0):
        #print(cars[column].name)
        if cars[column].dtype=='int32' or cars[column].dtype=='float64':
            cars[column].fillna(cars[column].mean(),inplace=True)
        
        else:
            print(cars[column].name)
            #cars[column].fillna(cars[column].astype(int).mean(),inplace=True)
    
```

```python
cars.dtypes

```

### Removing Columns with much lesser influence or values

```python
selectedColummns=["Make","Model","Height","Width","Length","Ex-Showroom_Price","Displacement","Cylinders","Valves_Per_Cylinder","Drivetrain","Cylinder_Configuration","Fuel_Tank_Capacity","Fuel_Type","Fuel_System","Emission_Norm"
,"Body_Type","Doors","City_Mileage","ARAI_Certified_Mileage","Kerb_Weight","Seating_Capacity","Wheelbase"]

for col in selectedColummns:
    print(col)
    print(cars[col].dtype)
```

```python
cars_selected
```

#### Filling NAN values

```python
for column in selectedColummns:
    if(cars_selected[column].isnull().sum()!=0):
        #print(cars[column].name)
        if cars_selected[column].dtype=='int32' or cars_selected[column].dtype=='float64':
            cars_selected[column].fillna(cars_selected[column].mean(),inplace=True)
        
        else:
            print(cars_selected[column].isnull().sum())
            
```

#### Bining 

```python
max(cars_selected["ARAI_Certified_Mileage"])
```

```python
mileage_columns=["ARAI_Certified_Mileage","City_Mileage"]
for col in mileage_columns:
    print(cars_selected[col])
    binwidth=(max(cars_selected[col])-min(cars_selected[col]))/4
    print(binwidth)
    bins=np.arange(min(cars_selected[col]),max(cars_selected[col]),binwidth)
    print(bins)
    group_names=["Low","Medium","High"]
    cars_selected[col+"_binned"]=pd.cut(cars_selected[col],bins,labels=group_names,include_lowest=True)
```

### Converting Categorical Features to Numerical Features

```python

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
catColumns=["Fuel_System","Cylinder_Configuration"]
for col in catColumns:
    cars_selected[col+"_converted"]=le.fit_transform(cars_selected[col])
for col in catColumns:
    cars_selected[col+"_converted"]=cars_selected[col+"_converted"].astype(int)
```

#### Selected Columns having more effect on Price

# Model Development

### Splitting the Data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df,cars_selected['Ex-Showroom_Price'],test_size=0.20,random_state=1)
```

```python
pd.set_option('display.max_rows', None)  # or 1000
cars_selected[["Fuel_System","Fuel_System_converted","Cylinder_Configuration","Cylinder_Configuration_converted"]].head(305)
```

### Fitting The Model to LinearRegression Model

```python
from sklearn.linear_model import LinearRegression
ln=LinearRegression()
ln.fit(x_train,y_train)
```

```python
ln.score(x_test,y_test)
```

```python
pred=ln.predict(x_test)

plt.figure(figsize=(16,8))


ax0=sns.distplot(cars_selected['Ex-Showroom_Price'],hist=False,color='g',label='Actual Value')
sns.distplot(pred,hist=False,color='g',label='Predicted Value',ax=ax0)

plt.title('Actual vs Predicted Value of Price')
plt.show()
```

### Fitting The Model toPolynomialRegression Model

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler


Input=[('scale',StandardScaler()),('polynomial',PolynomialFeatures(include_bias=False)),('model',LinearRegression())]
```

```python
pipe=Pipeline(Input)
pipe
```

```python
pipe.fit(x_train,y_train)
```

```python
pipe.score(x_test,y_test)
```

```python
pred=pipe.predict(x_test)

plt.figure(figsize=(20,8))


ax0=sns.distplot(cars_selected['Ex-Showroom_Price'],hist=False,color='g',label='Actual Value')
sns.distplot(pred,hist=False,color='g',label='Predicted Value',ax=ax0)

plt.title('Actual vs Predicted Value of Price')
plt.show()
```

### Fitting The Model to RandomForestRegression  Model

```python
model=RandomForestRegressor()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

### Tuning the Model

```python

for i in range(10,110,10):
    print(f"Trying model with {i} estimators...")
    model=RandomForestRegressor(n_estimators=i).fit(x_train,y_train)
    print(f"Model accuracy on test case: {model.score(x_test,y_test)*100:.2f}%")
```

```python
model=RandomForestRegressor(n_estimators=10).fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
